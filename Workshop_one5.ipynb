Создадим функцию, принимает на вход фрейм данных Pandas и возвращает подробное описание его характеристик. Она корректно обрабатывает возможные ошибки и предоставляет информативный вывод, даже если некоторые операции (например, describe для нечисловых столбцов) неприменимы. то есть выведем сразу статистику.

```python
def analyze_dataframe(df, df_name="DataFrame"):

    display(f"Analysis of {df_name}:\n")

    try:
        display("Head:\n", df.head(), "\n")
    except Exception as e:
        display(f"Error displaying head: {e}\n")

    try:
        display("Info:\n", df.info(), "\n")
    except Exception as e:
        display(f"Error displaying info: {e}\n")

    try:
        display("Describe:\n", df.describe(include='all'), "\n")  # Include all data types
    except Exception as e:
        display(f"Error displaying describe: {e}\n")

    try:
        duplicates = df.duplicated().sum()
        print(f"Number of duplicate rows: {duplicates}\n")
    except Exception as e:
        print(f"Error checking for duplicates: {e}\n")

    try:
        null_counts = df.isna().mean()#isnull().sum()
        display("Null value counts per column:\n", null_counts, "\n")
    except Exception as e:
        display(f"Error counting null values: {e}\n")

    try:
        display("Data types per column:\n", df.dtypes, "\n")
    except Exception as e:
        display(f"Error displaying data types: {e}\n")

    try:
        display("Shape:\n", df.shape, "\n")
    except Exception as e:
        display(f"Error getting shape: {e}\n")

analyze_dataframe(heart_train, "heart_train")
analyze_dataframe(heart_test, "heart_test")
```

```python
def rename_columns(df: pd.DataFrame) -> pd.DataFrame:
    """Переименовывает столбцы датасета в формат snake_case."""
    new_columns = {
        col: re.sub(r'[()]', '', col)  # Удаляем скобки
                   .lower()            # Приводим к нижнему регистру
                   .replace(' ', '_')  # Заменяем пробелы
                   .replace('-', '_')  # Заменяем дефисы
                   .strip('_')         # Убираем лишние подчеркивания
        for col in df.columns
    }
    return df.rename(columns=new_columns)

# Применяем функцию к датасетам
heart_train = rename_columns(heart_train)
heart_test = rename_columns(heart_test)

# Проверка совпадения столбцов
if sorted(heart_train.drop(columns='heart_attack_risk_binary').columns) != sorted(heart_test.columns):
    raise NameError('Столбцы датасетов не совпадают!')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def create_bar(df_list, col, title, legend):
    plt.figure(figsize=(15, 8))
    plt.grid(visible=True)
    plt.title(title)
    plt.xlabel('Категория')
    plt.ylabel('Частота, %')
    width = 0.2
    
    titles = df_list[0][col].value_counts().index
    x = np.arange(len(titles))
    
    for i, data in enumerate(df_list):
        plt.bar(
            x=x-(len(df_list)//2-i)*width, 
            width=width,
            height=data[col].value_counts(normalize=True)[titles] * 100,
            label=legend[i]
        )
    
    plt.legend(legend)
    plt.xticks(x, titles, rotation=45, ha='right')
    plt.tight_layout()

# Список всех числовых признаков
numeric_cols = [
    'age', 'cholesterol', 'heart_rate', 'exercise_hours_per_week',
    'bmi', 'sleep_hours_per_day', 'blood_sugar', 'ck_mb','troponin',
    'sedentary_hours_per_day', 'income', 'triglycerides'
]

# Список бинарных признаков
binary_cols = [
    'diabetes', 'family_history', 'smoking', 'obesity',
    'alcohol_consumption', 'previous_heart_problems', 'medication_use'
]

# Функция для преобразования типов
def convert_types(df):
    # Преобразуем числовые признаки в int
    for col in numeric_cols:
        # Сначала преобразуем в float, затем в int
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
    
    # Преобразуем бинарные признаки
    for col in binary_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
    
    return df

# Применяем преобразование
heart_train = convert_types(heart_train)
heart_test = convert_types(heart_test)

# Проверяем результаты
display("Типы данных в train после преобразования:")
display(heart_train.dtypes)
display("\nТипы данных в test после преобразования:")
display(heart_test.dtypes)

# Дополнительная проверка на наличие NaN
display("\nПроверка на NaN в train:")
display(heart_train.isnull().sum())
display("\nПроверка на NaN в test:")
display(heart_test.isnull().sum())
```

```python
def convert_float_to_int(df: pd.DataFrame, columns: list[str]) -> pd.DataFrame:
    """Функция для преобразования значений указанных колонок из float в int."""
    # Проверяем наличие NaN
    assert heart_train[columns].isna().sum().sum() == 0, "В данных присутствуют NaN"
    
    data = heart_train.copy()
    
    for col in columns:
        # Пропускаем колонку gender, так как она строковая
        if col != 'gender':
            # Очищаем данные и преобразуем
            data[col] = pd.to_numeric(
                data[col].astype(str).str.replace(',', '.').str.replace(' ', ''),
                errors='coerce'
            ).astype(int)
    
    # Проверки
    assert data.shape == heart_train.shape, "Размер DataFrame изменился"
    for col in columns:
        if col != 'gender':
            assert np.issubdtype(data[col].dtype, np.integer), f"Колонка {col} не преобразована"
    
    return data
```

```python
# Импортируем необходимые библиотеки
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Определяем препроцессинг для числовых признаков
numeric_features = [col['col'] for col in numeric_cols]
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Определяем препроцессинг для категориальных признаков
categorical_features = [col['col'] for col in category_cols]
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Объединяем препроцессинг
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# Определяем признаки для обучения
X_train = heart_train.drop(columns=['id', 'heart_attack_risk_binary'])
y_train = heart_train['heart_attack_risk_binary']

# Для тестового датасета убираем только id
X_test = heart_test.drop(columns=['id'])

# Создаем пайплайн
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])

# Параметры для поиска
param_grid = {
    'classifier__C': [0.1, 1, 10],
    'classifier__solver': ['liblinear', 'saga']
}

# Настройка GridSearch
grid_search = GridSearchCV(
    pipe, 
    param_grid, 
    cv=5,
    scoring='f1',
    n_jobs=-1
)

# Обучение модели
grid_search.fit(X_train, y_train)

# Предсказания на тестовом датасете
y_pred = grid_search.predict(X_test)

# Сохранение результатов
predictions = pd.DataFrame({
    'id': heart_test['id'],
    'heart_attack_risk_binary': y_pred
})

print("Лучшие параметры модели:")
print(grid_search.best_params_)

# Оценка на тренировочной выборке (для проверки переобучения)
y_train_pred = grid_search.predict(X_train)
print("\nМетрики на тренировочной выборке:")
print(f'F1-score: {f1_score(y_train, y_train_pred):.3f}')
print(f'ROC-AUC: {roc_auc_score(y_train, grid_search.predict_proba(X_train)[:,1]):.3f}')
print(f'Accuracy: {accuracy_score(y_train, y_train_pred):.3f}')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Создаем пайплайн с Random Forest
rf_pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Параметры для поиска
rf_param_grid = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5],
    'classifier__class_weight': ['balanced', None]
}

# Настройка GridSearch
rf_grid_search = GridSearchCV(
    rf_pipe, 
    rf_param_grid, 
    cv=5,
    scoring='f1',
    n_jobs=-1
)

# Обучение модели
rf_grid_search.fit(X_train, y_train)

# Предсказания
y_pred_rf = rf_grid_search.predict(X_test)

# Оценка модели
print("Лучшие параметры модели RF:")
print(rf_grid_search.best_params_)

y_train_pred_rf = rf_grid_search.predict(X_train)
print("\nМетрики на тренировочной выборке RF:")
print(f'F1-score: {f1_score(y_train, y_train_pred_rf):.3f}')
print(f'ROC-AUC: {roc_auc_score(y_train, rf_grid_search.predict_proba(X_train)[:,1]):.3f}')
print(f'Accuracy: {accuracy_score(y_train, y_train_pred_rf):.3f}')
print("\nClassification Report:")
print(classification_report(y_train, y_train_pred_rf))
```

* Корреляционный анализ данных
Целевым признаком является Покупательская_активность и вот список полей, которые имеют хоть корреляцию с ним: Маркет_актив_6_мес, Акционные_покупки, Средний_просмотр_категорий_за_визит, Неоплаченные_продукты_штук_квартал, Страниц_за_визит, Выручка_препредыдущий_месяц, минут_предыдущий_месяц, минут_текущий_месяц.
Так же график показал сильную зависимость между целевым признаком и id так что в дальнейшем при подготовке данных к обучению модели признак идентификатор был удален. Была выявлена высокая корреляция между «выручка_текущий_мес» и «выручка_предыдущий_мес», что  указывает на мультиколлинеарность, которая может повлиять на стабильность оценок параметров модели Так как значение корреляции меньше 0.9. Поэтому были удалены признаки «выручка_предыдущий_мес» и «минут_предыдущий_месяц». Сильной корреляции между другими признаками не было выявлено поэтому все остальные признаки были оставленны в датасете.

* Обучение модели
С использованием пайпланов из библиотеки sklearn были обучены модели KNeighborsClassifier(), DecisionTreeClassifier(), LogisticRegression() и SVC(). При обучении моделей средстави пайплана преебирались некоторое количество гиперпараметров моделей с целью нахождения модели, которая даст лучший результат.
По результатам обучения лучшей стала модель LogisticRegression(C=2, penalty='l1', random_state=42, solver='liblinear'). Метрика roc_auc для тренеровочной выборки 0.90 для тестовой 0.89. Удивительно, но значения метрики имеют такой низкий разброс. И даже наличие дисбаланса в целевом признаке не помешало получить значение метрики f1=0.90.

* Анализ важности признаков
Если коротко, то самыми важными оказались признаки связанные с количеством времени, которое покупатель проводит на меркет плейсе. А наименее важными, категории товаров, котоыре интересовали пользователи и комменикации с пользователем.
Кажется что все эти маркетинговые завлекаловки и спам на почту не помогают маркетплейсу. Важно лишь заставить пользователя проводить на сайте как можно больше времени и продажи выростут.

* Сегментация пользователей
Для анализа пользователей был выбран сегмент пользователей часто покупающих по скидке и в то же время имеющийх высокую вероятность снижения покупательской активности. Мое предложение по увеличение актвности этого сегмента покупателей это таргетированная настройка главной страницы маркетплейса. А так же более частое обновление скидок на главной странице. Пользователь не хочет ходить по страницам, он хочет видеть все предложения в одном месте, а не терять свое время на перелистование страниц и читание электроной почты и смс.

Потенциал для роста.Что это значит для бизнеса:
Большинство клиентов активно пользуются платформой(так покупательская активность доходит в среднем 98.6%), эти покупатели смотрят несколько категорий товаров, оставляют товары в корзине и имеют стабильные траты. Но есть проблема: очень высокая вероятность того, что клиенты снизят активность (86.5% в среднем). Тогда нужно срочно принимать меры для удержания.
